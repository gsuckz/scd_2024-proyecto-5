```python
import numpy as np
import matplotlib.pyplot as plt

def coseno_alzado_espectro(f, T=1, beta=0.25):
    # Definicin de la funcin del coseno alzado
    h = np.zeros_like(f)
    for i in range(len(f)):
        if abs(f[i]) <= (1-beta)/(2*T):
            h[i] = T

        elif (1 - beta)/(2*T) < abs(f[i]) <= (1+beta)/(2*T):

            h[i]= 0.5*(1+np.cos(np.pi*T*(abs(f[i])-(1-beta)/(2*T))/beta))
        else:
            h[i] = 0
            #print("0")
    return h

# Definir el rango de t
f = np.linspace(-1, 1, 10000)


# Valores de beta para superponer
betas = [0.01, 0.25, 0.5, 0.75, 1.0]
colors = ['b', 'g', 'r', 'c', 'm']

# Graficar las curvas de coseno alzado
plt.figure(figsize=(10, 6))

for beta, color in zip(betas, colors):
    pulso_coseno = coseno_alzado_espectro(f, T=1, beta=beta)
    plt.plot(f, pulso_coseno, label=f' = {beta}', color=color)

# Personalizar la grfica
plt.title('Curvas de Coseno Alzado para Diferentes Valores de ')
plt.xlabel('f')
plt.ylabel('H(f)')
plt.grid()
plt.legend()
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.xlim(-1, 1)
plt.ylim(-0.1, 1.1)
plt.show()
```
